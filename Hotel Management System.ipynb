```python
def date_Format(a):
    if "-" in a:
        s=a.split("-")
        s.reverse()
        s="-".join(s)
        return s
    elif "/" in a:
        s=a.split("/")
        s.reverse()
        s="-".join(s)
        return s
```

```python
def check_date(check_in,check_out):
    if "-" in check_in:
        c1=check_in.split("-")
        c2=check_out.split("-")
        a=0
        if c1[0]<=c2[0]:
            if c1[1]<=c2[1]:
                if c1[2]<=c2[2]:
                    z=True
                    return z
        if a==0:
            z=False
            return z
    elif "/" in check_in:
        c1=check_in.split("/")
        c2=check_out.split("/")
        a=0
        if c1[0]<=c2[0]:
            if c1[1]<=c2[1]:
                if c1[2]<=c2[2]:
                    z=True
                    return z
        if a==0:
            z=False
            return z
        
```

```python
def check_date1(c1,c2):
    if c1 <= c2:
        x=True
        return x
    else:
        x=False
        return x
```

```python
def date_ch_after_form1(ch_in,i,j):
    ch_in=date.fromisoformat(ch_in)
    if ch_in<=l[i][j]:
        x=True
        return x
    else:
        x=False
        return x
```

```python
def date_ch_after_form2(ch_out,i,j):
    ch_out=date.fromisoformat(ch_out)
    if ch_out>=l[i][j]:
        x=True
        return x
    else:
        x=False
        return x
```

```python
def no_of_days(c_in,c_out):
    date1=date.fromisoformat(c_in)
    date2=date.fromisoformat(c_out)
    gap=str(date1-date2)
    for i in range(len(gap)):
        if gap[i].isdigit():
            g=int(gap[i])
            break
    g+=1
    return g
```

```python
def no_of_days1(c_in,c_out):
    gap=str(c_in-c_out)
    for i in range(len(gap)):
        if gap[i].isdigit():
            g=int(gap[i])
            break
    g+=1
    return g
```
